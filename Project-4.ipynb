```python
def MonteCarlo(f, samples):
    N = len(samples)
    return 1/float(N)*sum([f(e) for e in samples])

omega1 = MonteCarlo(lambda x: x[0], samples)
omega2 = MonteCarlo(lambda x: x[1], samples)
omega3 = MonteCarlo(lambda x: x[2], samples)
print ('o1 = {0}, o2 = {1}, o3 = {2}'.format(omega1, omega2, omega3))
```

```python
# Comparing model to data
Xval = np.arange(0,11)
RegressionEquation = [omega1 + omega2 * np.sin(x) + omega3 * np.cos(x) for x in Xval]
# w1 + (w2 * sin(x) + w3 * cos(x))

X = [x[0] for x in data]
T = [x[1] for x in data]
fig, ax = plt.subplots(1,1, figsize = (5,3))
ax.scatter(T,X)
ax.plot(Xval, RegressionEquation, label ='Regression', color = 'red')
ax.set_xlabel('t')
ax.set_ylabel('$X_t$')
ax.legend(loc = 'best', fontsize='small')
ax.grid()
```
